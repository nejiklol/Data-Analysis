Импортируем датасет MNIST.

```python
from keras.datasets import mnist
(X_train, y_train), (X_pred, y_pred) = mnist.load_data()
```

```python
dim = 784 # 28*28 
X_train = X_train.reshape(len(X_train), dim)

from sklearn.decomposition import PCA

pca = PCA(n_components=100, svd_solver='full')
modelPCA = pca.fit(X_train) 

X_train = modelPCA.transform(X_train)

explained_variance = np.cumsum(pca.explained_variance_ratio_)
M, meaning = 0, 0
accuracy = float(input("Твоя точность: "))
while meaning <= accuracy:
    meaning = explained_variance[M]
    M += 1
print("Количество элементов: ", M)
```

Определим количество элементов, соответствующих минимуму доли объясненной дисперсии:

```python
from keras.datasets import mnist
(X_train, y_train), (X_pred, y_pred) = mnist.load_data()

dim = 784 # 28*28 
X_train = X_train.reshape(len(X_train), dim)

from sklearn.decomposition import PCA

pca = PCA(n_components=M, svd_solver='full')
modelPCA = pca.fit(X_train) 

X_train = modelPCA.transform(X_train)
from sklearn.model_selection import train_test_split
rs = int(input("Твой random_state: "))
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.3, random_state=rs)
print("Выборочное среднее нулевого столбца: {:.2f}".format(X_train.T[0].mean()))
```

Подключаем библиотеки для обучения многоклассового классификатора.

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

Случайное дерево:

```python
i = 0
tree = RandomForestClassifier(criterion='gini', min_samples_leaf=10, max_depth=20, n_estimators=10, random_state=rs)
clf1 = OneVsRestClassifier(tree).fit(X_train, y_train)
y_pred = clf1.predict(X_test)
CM = confusion_matrix(y_test, y_pred)
CM[i][i]
```

Логистическая регрессия:

```python
i = 1
tree = LogisticRegression(solver='lbfgs', random_state=rs)
clf2 = OneVsRestClassifier(tree).fit(X_train, y_train)
y_pred = clf2.predict(X_test)
CM = confusion_matrix(y_test, y_pred)
CM[i][i]
```

Дерево принятия решений:

```python
i = 7
tree = DecisionTreeClassifier(criterion='gini', min_samples_leaf=10, max_depth=20, random_state=rs)
clf3 = OneVsRestClassifier(tree).fit(X_train, y_train)
y_pred = clf3.predict(X_test)
CM = confusion_matrix(y_test, y_pred)
CM[i][i]
```
